```python
import numpy as np

u = np.array([8,2,5])
v_action_movie = np.array([9,1,2])
v_comedy_movie = np.array([1,9,8])
v_drama_movie = np.array([7,2,6])

def cosine_similarity_score(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))

action_movie_score = ("Action Movie", cosine_similarity_score(u, v_action_movie))
comedy_movie_score = ("Comedy Movie", cosine_similarity_score(u, v_comedy_movie))
drama_movie_score = ("Drama Movie", cosine_similarity_score(u, v_drama_movie))

print(f"{action_movie_score[0]}: {action_movie_score[1]}")
print(f"{comedy_movie_score[0]}: {comedy_movie_score[1]}")
print(f"{drama_movie_score[0]}: {drama_movie_score[1]}")
print("The best match is:")
best_score = max(action_movie_score, comedy_movie_score, drama_movie_score, key=lambda x: x[1])
print(f"{best_score[0]}: {best_score[1]}")
```

Matrix-based cosine similarity

```python
import numpy as np

user = np.array([8,2,5])
movies = np.array([
    [9,1,2],
    [1,9,8],
    [7,2,6]
    ])
movie_genre = ['Action Movie', 'Comedy Movie', 'Drama Movie']


def cosine_similarity_score(user, movies):
    return np.dot(movies, user) / (np.linalg.norm(movies, axis=1) * np.linalg.norm(user))  

user_similarity = cosine_similarity_score(user, movies)
for i in range(len(movie_genre)):
    print(f"{movie_genre[i]}: {user_similarity[i]}")
print("The best match is:")
movie_index = user_similarity.argmax()
print(f"{movie_genre[movie_index]}: {user_similarity[movie_index]}")

```
