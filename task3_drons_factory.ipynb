```python
import numpy as np

A = np.array([
    [4,4,6],
    [1,1,2],
    [1,2,4]
    ])
drone_names = ['Scout', 'Kamikaze', 'Cargo']

b = np.array([460,130,240])
determinant = np.linalg.det(A)
if determinant != 0:
    print(f"Determinant is {determinant} which is not equal to 0, the system has a unique solution.")
    x = np.linalg.solve(A, b)
    for i in range(len(drone_names)):
        print(f"Totally factory created {x[i]} {drone_names[i]}")
    print(f"Total number of drones created: {np.sum(x)}")
    print("Let's check if the solution is correct:")
    print("Materials were used:")
    used_materials = np.dot(A, x)
    print(used_materials)
    print(f"Used materials are equal to the total materials: {used_materials == b}")
else:
    print("The system has no unique solution.")
```
