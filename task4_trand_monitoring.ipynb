```python
import numpy as np

t = np.array([1, 2, 3, 4, 5])
y = np.array([22, 28, 37, 45, 53])

A = np.column_stack([t, np.ones(len(t))])
result = np.linalg.lstsq(A, y, rcond=None)
k, b = result[0]

print(f"k = {k}, b = {b}, The line equation is y = {k}x + {b:.2f}")
print(f"Let's predict the server loading at t = 6: k * 6 + b = {k * 6 + b:.2f}%")
user_input = input("Enter the time when you want to server loading: ")
if user_input.strip() == "":
    print("No input provided. Exiting.")
else:
    try:
        current_t = float(user_input)
        current_y = k * current_t + b
        if current_y > 100:
            print(f"Warning: Predicted value {current_y:.2f}% exceeds 100%, so probably your server will be already down")
            current_y = 100
        elif current_y < 0:
            print(f"Warning: Predicted value {current_y:.2f}% is negative")
            current_y = 0
        else:
            print(f"The server loading at time {current_t} is {current_y:.2f}%")
    except ValueError:
        print("Invalid input. Please enter a number.")
```
